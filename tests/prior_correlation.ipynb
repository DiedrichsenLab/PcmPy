## Imposing a prior on a correlation model

```python
# Visualize a log-normal prior on the fisher-z value
z = np.random.normal(0,2,1000000)
r = (exp(2*z)-1)/(exp(2*z)+1)
sb.histplot(r,binrange=(-1,1),bins=1000)
```
